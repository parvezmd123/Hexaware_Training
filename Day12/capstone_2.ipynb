```python
# 34.
inventory_transformed_df = inventory_with_products_df.withColumn("stock_status",
    when(col("stock_quantity") <= col("reorder_level"), "Reorder Required").otherwise("Sufficient Stock"))
inventory_transformed_df.show(5)

# 35.
products_transformed_df = products_with_suppliers_df.withColumn("price_category",
    when(col("unit_price") >= 50000, "Premium") \
    .when(col("unit_price") >= 10000, "Standard") \
    .otherwise("Budget"))
products_transformed_df.show(5)

# 36. Create sales_transformed_df with correct, unambiguous column names and new transformations
sales_transformed_df = retail_sales_df.select(
    col("sale_id"),
    col("store_id"),
    col("product_id"),
    col("sale_date"), # Already date type from retail_sales_df
    col("quantity_sold"),
    col("sale_amount"),
    col("payment_m"),
    col("sale_data_quality_status"),
    col("store_name"),
    col("store_city"),
    col("state"),
    col("store_type"),
    col("manager_name"),
    col("product_name"),
    col("category"),
    col("brand"),
    col("unit_price"),
    col("product_data_quality_status"),
    col("supplier_city"),
    col("rating"),
    col("supplier_name"),
    col("phone"),
    col("email"),
    retail_sales_df.supplier_id # Explicitly reference supplier_id from retail_sales_df
)

sales_transformed_df = sales_transformed_df.withColumn("revenue_category",
    when(col("sale_amount") >= 50000, "High Revenue") \
    .when(col("sale_amount") >= 15000, "Medium Revenue") \
    .otherwise("Low Revenue")) \
    .withColumn("sale_month", month(col("sale_date"))) \
    .withColumn("sale_year", year(col("sale_date")))

sales_transformed_df.show(5)
```

### Part 7: Window Functions

```python
from pyspark.sql.window import Window

# 56. Rank products by revenue
print("\nProducts Ranked by Revenue:")
window_spec_product_rank = Window.orderBy(col("total_revenue").desc())

ranked_products = sales_transformed_df.groupBy("product_id", "product_name")\
                                        .agg(sum("sale_amount").alias("total_revenue"))\
                                        .withColumn("revenue_rank", rank().over(window_spec_product_rank))
ranked_products.show()

# 57. Rank stores by revenue
print("\nStores Ranked by Revenue:")
window_spec_store_rank = Window.orderBy(col("total_revenue").desc())

ranked_stores = sales_transformed_df.groupBy("store_id", "store_name")\
                                      .agg(sum("sale_amount").alias("total_revenue"))\
                                      .withColumn("revenue_rank", rank().over(window_spec_store_rank))
ranked_stores.show()
```

```python
# 58. Rank products by revenue within category
print("\nProducts Ranked by Revenue within Category:")
window_spec_category_product_rank = Window.partitionBy("category").orderBy(col("total_revenue").desc())

ranked_products_in_category = sales_transformed_df.groupBy("category", "product_id", "product_name")\
                                                      .agg(sum("sale_amount").alias("total_revenue"))\
                                                      .withColumn("category_revenue_rank", rank().over(window_spec_category_product_rank))
ranked_products_in_category.show()

# 59. Find top product in each category
print("\nTop Product in Each Category:")
ranked_products_in_category.filter(col("category_revenue_rank") == 1).show()

# 60. Find top 3 products in each category
print("\nTop 3 Products in Each Category:")
ranked_products_in_category.filter(col("category_revenue_rank") <= 3).show()
```

```python
# 61. Find top store by state
print("\nTop Store by State:")
window_spec_state_store_rank = Window.partitionBy("state").orderBy(col("total_revenue").desc())

ranked_stores_in_state = sales_transformed_df.groupBy("state", "store_id", "store_name")\
                                                 .agg(sum("sale_amount").alias("total_revenue"))\
                                                 .withColumn("state_revenue_rank", rank().over(window_spec_state_store_rank))
ranked_stores_in_state.filter(col("state_revenue_rank") == 1).show()
```

```python
# 62. Calculate running total revenue by sale_date
print("\nRunning Total Revenue by Sale Date:")
window_spec_running_total = Window.orderBy("sale_date")

running_total_revenue = sales_transformed_df.groupBy("sale_date")\
                                            .agg(sum("sale_amount").alias("daily_revenue"))\
                                            .withColumn("running_total_revenue", sum("daily_revenue").over(window_spec_running_total))
running_total_revenue.show()

# 63. Use lag to compare current day and previous day sales
print("\nDaily Sales with Previous Day's Sales (Lag):")
window_spec_lag = Window.orderBy("sale_date")

sales_with_lag = sales_transformed_df.groupBy("sale_date")\
                                    .agg(sum("sale_amount").alias("daily_revenue"))\
                                    .withColumn("previous_day_revenue", lag("daily_revenue", 1).over(window_spec_lag))
sales_with_lag.show()

# 64. Use lead to show next sale amount
print("\nSales with Next Sale Amount (Lead):")
window_spec_lead = Window.partitionBy("product_id").orderBy("sale_date")

sales_with_lead = sales_transformed_df.withColumn("next_sale_amount", lead("sale_amount", 1).over(window_spec_lead))
sales_with_lead.select("sale_id", "product_name", "sale_date", "sale_amount", "next_sale_amount").show()
```

```python
# 65. Find products whose revenue increased compared to previous sale
print("\nProducts with Increased Revenue Compared to Previous Sale:")
window_spec_prev_sale = Window.partitionBy("product_id").orderBy("sale_date")

products_revenue_increase = sales_transformed_df.withColumn("previous_sale_amount", lag("sale_amount", 1).over(window_spec_prev_sale))\
                                                 .filter(col("sale_amount") > col("previous_sale_amount"))
products_revenue_increase.select("product_name", "sale_date", "previous_sale_amount", "sale_amount").show()
```

```python
# 78. Create incremental sales file for March 2026
incremental_sales_data = [
    ("SA1016", "S101", "P101", "2026-03-01", 1, 65000, "Card"),
    ("SA1017", "S102", "P102", "2026-03-01", 2, 50000, "UPI"),
    ("SA1018", "S103", "P106", "2026-03-02", 3, 13500, "Cash")
]

incremental_sales_schema = StructType([
    StructField("sale_id", StringType(), True),
    StructField("store_id", StringType(), True),
    StructField("product_id", StringType(), True),
    StructField("sale_date", StringType(), True),
    StructField("quantity_sold", IntegerType(), True),
    StructField("sale_amount", IntegerType(), True),
    StructField("payment_m", StringType(), True)
])

incremental_sales_df_raw = spark.createDataFrame(incremental_sales_data, schema=incremental_sales_schema)
incremental_sales_df_raw.write.mode("overwrite").csv("incremental_sales.csv", header=True)
print("Incremental sales file created.")
```

```python
# 79. Read incremental sales file
incremental_sales_df = spark.read.csv("incremental_sales.csv", header=True, inferSchema=True)
print("Incremental sales file read.")
incremental_sales_df.show()

# 80. Append incremental sales to Silver sales Parquet
# First, apply the same cleaning steps to the incremental data as the original sales_df
incremental_sales_cleaned_df = incremental_sales_df.na.fill({'sale_amount': 0, 'payment_m': 'Not Provided'}) \
                                                  .withColumn("data_quality_status", lit("Good")) # Assuming incremental data is clean or handled

# Now append to the existing silver sales parquet
incremental_sales_cleaned_df.write.mode("append").parquet("silver/sales.parquet")
print("Incremental sales appended to Silver sales Parquet.")
```

```python
# Reload the full sales dataset from Silver to include incremental data
full_sales_silver_df = spark.read.parquet("silver/sales.parquet")

# Re-apply transformations to the full sales data to get an updated sales_transformed_df
# Join with stores and products to maintain full context
full_sales_transformed_df = full_sales_silver_df \
    .join(stores_silver_df, on="store_id", how="left") \
    .join(products_silver_df, on="product_id", how="left") \
    .join(suppliers_flattened_df, on="supplier_id", how="left") \
    .withColumn("revenue_category",
        when(col("sale_amount") >= 50000, "High Revenue") \
        .when(col("sale_amount") >= 15000, "Medium Revenue") \
        .otherwise("Low Revenue")) \
    .withColumn("sale_date", to_date(col("sale_date"))) \
    .withColumn("sale_month", month(col("sale_date"))) \
    .withColumn("sale_year", year(col("sale_date")))

print("Full sales data reloaded and transformed after incremental load.")
full_sales_transformed_df.show(5)
```
