```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Day 1") \
    .getOrCreate()
```

```python
data = [
    (1, "Rahul Sharma", "Hyderabad", "Electronics", "Laptop", 1, 65000, "Delivered"),
    (2, "Priya Reddy", "Bangalore", "Electronics", "Mobile", 2, 25000, "Delivered"),
    (3, "Amit Kumar", "Mumbai", "Furniture", "Chair", 4, 3500, "Pending"),
    (4, "Sneha Patel", "Chennai", "Electronics", "Tablet", 1, 30000, "Cancelled"),
    (5, "Farhan Ali", "Delhi", "Furniture", "Table", 2, 12000, "Delivered"),
    (6, "Neha Singh", "Hyderabad", "Fashion", "Shoes", 3, 2500, "Delivered"),
    (7, "Arjun Verma", "Pune", "Fashion", "Watch", 1, 8000, "Pending"),
    (8, "Meera Nair", "Bangalore", "Electronics", "TV", 1, 45000, "Delivered"),
    (9, "Kiran Rao", "Hyderabad", "Furniture", "Desk", 1, 15000, "Delivered"),
    (10, "Ravi Kumar", "Mumbai", "Electronics", "Mobile", 1, 28000, "Delivered")
]
```

```python
columns = [
"order_id",
"customer_name",
"city",
"category",
"product",
"quantity",
"price",
"order_status"
]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
df_total = df.withColumn("total_amount", col("quantity") * col("price"))
df_total.show()
```

```python
df_discount = df_total.withColumn("discount", col("total_amount") * 0.10)
df_discount.show()
```

```python
df_final = df_discount.withColumn(
    "final_amount",
    col("total_amount") - col("discount")
)
df_final.show()
```

```python
df_category = df_total.withColumn(
    "order_value_category",
    when(col("total_amount") >= 50000, "High")
    .when(col("total_amount") >= 20000, "Medium")
    .otherwise("Low")
)

df_category.show()
```

```python
df_final.filter(
    col("order_status") == "Delivered"
).select(
    "order_id",
    "customer_name",
    "final_amount"
).show()
```

```python
df_final.filter(
    col("order_status") == "Delivered"
).groupBy(
    "city"
).sum(
    "final_amount"
).show()
```

```python
final_report = df.withColumn(
    "total_amount",
    col("quantity") * col("price")
).withColumn(
    "discount",
    col("quantity") * col("price") * 0.10
).withColumn(
    "final_amount",
    (col("quantity") * col("price")) -
    ((col("quantity") * col("price")) * 0.10)
).select(
    "order_id",
    "customer_name",
    "city",
    "category",
    "product",
    "quantity",
    "price",
    "total_amount",
    "discount",
    "final_amount",
    "order_status"
)

final_report.show()
```
