```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Day 1") \
    .getOrCreate()
```

```python
spark
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]
```

```python
columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
from pyspark.sql.functions import col, when

df_category = df.withColumn(
    "salary_category",
    when(col("salary") >= 85000, "High")
    .when(col("salary") >= 75000, "Medium")
    .otherwise("Low")
)
df_category.show()
```

```python
df_category.groupBy("salary_category").count().show()
```

```python
from pyspark.sql.functions import col
final_report=df.withColumn("bonus",col("salary")*0.10) \
               .withColumn("annual_salary",col("salary")*12) \
               .select(
                   "employee_id",
                   "name",
                   "department",
                   "salary",
                   "bonus",
                   "annual_salary"
               )
final_report.show()
```
